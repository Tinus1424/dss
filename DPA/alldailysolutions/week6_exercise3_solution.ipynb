For this exercise we will be working again with the Iris dataset.

- First, load the dataset from `iris.csv` file and convert it to a numpy array (you may leave out the fifth column). 
- Create a boxplot that shows the petal length data of each iris species separately. Make sure the means are also visible, and add labels and a title.

```python
%matplotlib inline

import numpy
import pandas
import matplotlib.pyplot as plt
```

```python
# Loading and preparing the dataset
iris = pandas.read_csv("iris.csv", index_col=0).to_numpy()
iris = iris[:, :-1] #remove the fifth column
```

```python
#subsetting the dataset
seto = iris[:50, :]
vers = iris[50:100, :]
virg = iris[100:, :]

pet_len_set = seto[:, 2]
pet_len_ver = vers[:, 2]
pet_len_vir = virg[:, 2]
```

```python
plt.figure(1)
plt.boxplot(x = [pet_len_set, pet_len_ver, pet_len_vir], 
            labels = ["iris setosa", "iris versicolor", "iris virginica"],
            showmeans = True)
plt.ylabel("petal length (cm)")
plt.title("Petal lengths in different iris species", fontsize = 15)
```
