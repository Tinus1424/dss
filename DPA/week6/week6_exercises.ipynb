
First, load the dataset from iris.csv file and convert it to a numpy array (you may leave out the fifth column).

Now, create a figure with 4 subplots. Each subplot should be a scatterplot, plotting the data of one column in the iris dataset against the data of another column (e.g. petal length against sepal width). 

Many combinations are possible, but you may stick to 4 for this exercise.

Make sure the plot is annotated properly adding titles, x and y labels.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
iris = pd.read_csv("../data/iris.csv")
iris = np.asarray(iris)
df = iris[:,1 : 5]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 6), 
                         sharex=True)
fig.suptitle("COOL PLOTS WITH COOL DOTS")
ax[0, 0].scatter(df[:, 0], df[:, 1], c = "blue")
ax[0, 0].set_xlabel("Sepal Length")
ax[0, 0].set_ylabel("Sepal Width")
ax[0, 1].scatter(df[:, 0], df[:, 2], marker = "$4$")
ax[0, 1].set_xlabel("Sepal Length")
ax[0, 1].set_ylabel("Petal Length")
ax[1, 0].scatter(df[:, 0], df[:, 3])
ax[1, 0].set_xlabel("Sepal Length")
ax[1, 0].set_ylabel("Petal Width")
ax[1, 1].scatter(df[:, 1], df[:, 2])
ax[1, 1].set_xlabel("Sepal Width")
ax[1, 1].set_ylabel("Petal Length")
```
